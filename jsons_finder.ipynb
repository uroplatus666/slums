```python
import geopandas as gpd
import folium
import numpy as np
from shapely.geometry import box
import os

mumbai = gpd.read_file('Границы/mumbai_boundary.gpkg')
nd = gpd.read_file('Границы/new_delhi_boundary.gpkg')
bangalore = gpd.read_file('Границы/bangalore_boundary.gpkg')
```

```python
directory_path = 'data'

geojson_files = [f for f in os.listdir(directory_path) if f.endswith('.geojson')]


for file in geojson_files:
    gdf = gpd.read_file('data/'+file)
    mumbai_poly = mumbai.geometry.iloc[0]
    sindex = gdf.sindex
    xmin, ymin, xmax, ymax = mumbai_poly.bound
    possible_idxs = list(sindex.intersection((xmin, ymin, xmax, ymax)))
    
    if not possible_idxs:
        print("Нет ни одного полигона, даже по bbox, близко к Mumbai.")
    else:
        for idx in possible_idxs:
            poly = gdf.geometry.iloc[idx]
            if poly.within(mumbai_poly):
                print(f"Полигон с индексом {idx} полностью внутри Mumbai.")
                print(f'Посмотри этот файл: {file}!')
                break

# Полигон с индексом 225067 полностью внутри Mumbai.
# Посмотри этот файл: 123300311.geojson!
```
